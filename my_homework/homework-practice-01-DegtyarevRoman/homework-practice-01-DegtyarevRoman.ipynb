### Задача 2: кросс-валидация для исходных данных
__(0.5 балла)__

Загрузите файл scores.csv и протестируйте, как линейная регрессия предсказывает ответ сейчас (с помощью кросс-валидации).

*Hint: воспользуйтесь sklearn.linear_model и sklearn.model_selection.*

```python
# Your code here
# ...
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
y = pd.read_csv('scores.csv', header = None)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.33, random_state=42)
```

```python
print("Linear regression score")
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print("MSE:\t"+str(mean_squared_error(y_test, y_predict)))
print("r2:\t" +str(r2_score(y_test, y_predict)))
```

```python
print("ElasticNet score")
model = ElasticNet(alpha=0.5).fit(X_train, y_train)
y_predict = model.predict(X_test)
print("MSE:\t"+str(mean_squared_error(y_test, y_predict)))
print("r2:\t" +str(r2_score(y_test, y_predict)))
```
